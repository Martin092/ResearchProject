```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def normal(x, mu=0, v=1.2):
    c = 1 / np.sqrt(2 * np.pi * v ** 2)
    return c * np.exp(-((x - mu)**2) / (2 * v**2))
```

```python
def plot_dist(f):
    b = 5
    x = np.linspace(-b, b, 1000)
    y = f(x)
    plt.plot(x, y)
    plt.fill_between(x, y, 0, alpha=0.5)
    plt.axvline(x=0, color='r', linestyle=':', label='Mean')
```
